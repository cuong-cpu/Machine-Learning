### Pyplot API vs Object-Oriented (OO) API

Pyplot API -> Quickly

```python
x = np.linspace(0,10)
plt.plot(x,np.sin(x), color='blue',linestyle ='dashed',label="Sin(x)")
plt.plot(x,np.cos(x),color='red',label='Cos(x)')
plt.xlabel("bien x")
plt.ylabel("F(x)")
plt.axis("equal")
# plt.xlim([0,4])
# plt.ylim([-0.7,0.6])
plt.legend()
```

Object-Oriented API

```python
x = [1,2,3,4]
y = [11,22,33,44]

fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title='A simple plot',
       xlabel="x-axis",
       ylabel="y-axis");
```

### 2. Most common types of Matplotlib plots

Line

```python
x = np.linspace(0,10,100)
x[:5]
```

```python
# OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x,x**3)
```

Scatter

```python
plt.scatter(x,np.exp(x));
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s = sizes, c = colors,cmap = 'viridis',alpha=0.3)
fig.colorbar(img1);
```

3.Bar

```python
soft_drink_prices = {"Coke":10,
                     "Pessi":12,
                     "Sprite":8}
fig, ax =plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title="Bach Hoa Xanh", ylabel="Price")
```
